```python
plankton_source = dataset.g_mole_concentration_of_phytoplankton_expressed_as_carbon_in_sea_water.g_mole_concentration_of_phytoplankton_expressed_as_carbon_in_sea_water[:,0,:,:]
lat = dataset.g_mole_concentration_of_phytoplankton_expressed_as_carbon_in_sea_water.lat0[:]
lon = dataset.g_mole_concentration_of_phytoplankton_expressed_as_carbon_in_sea_water.lon0[:]
ts = dataset.g_mole_concentration_of_phytoplankton_expressed_as_carbon_in_sea_water.time0[:]
x, y = np.meshgrid(lon, lat)
```

I will use matplotlib to build a map of a phytoplankton concentration. The <i>plankton_source</i> variable represents the 4-D array of phytoplankton concentration with time, depth and lat/lon dimensions. From beautiful visualization let's turn into some analysis and calculate a mean value of a phytoplankton concentration over this region. 

```python
import matplotlib.dates as mdates
plt.close('all')
fig = plt.figure(figsize=(10,5))

fig.suptitle('Mean phytoplankton concentration over the North Atlantic (2012)', fontsize=14, fontweight='bold')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m'))
ax = fig.add_subplot(111)

ax.set_xlabel('Month')
ax.set_ylabel('mmol/m**3')

means = list()
times = list()
for i, t in enumerate(ts):           
    dt = datetime.datetime.fromtimestamp(t)	
    plankton = plankton_source[i][0][::-1,::]        
    plankton = np.ma.masked_where(isnan(plankton), plankton)
    means.append(np.mean(np.ma.fix_invalid(plankton)))
    times.append(dt)
ax.plot(times, means)
plt.show()
```

There is no something we did not expect: middle of summer is a peak for the phytoplankton bloom. Hovewer this is not always the same for internal seas due different salinity, rivers runoff and other reasons causing plankton species diversity. Thus it is known that a peak of a bloom for the Baltic Sea falls on March - April. It can be easily checked by reducing the same dataset with the boundaries of the Baltic Sea. 

```python
lat_n = 66.3
lat_s = 53
lon_w = 12.5
lon_e = 31.1
latindexes = list()
lonindexes = list()
for i, la in enumerate(lat):
    if la < lat_n and la > lat_s:
        latindexes.append(i)
for i, lo in enumerate(lon):
    if lo < lon_e and lo > lon_w:
        lonindexes.append(i) 
        
plt.close('all')
fig = plt.figure(figsize=(10,5))

fig.suptitle('Mean phytoplankton concentration over the Baltic Sea (2012)', fontsize=14, fontweight='bold')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m'))
ax = fig.add_subplot(111)

ax.set_xlabel('Month')
ax.set_ylabel('mmol/m**3')

means = list()
times = list()
for i, t in enumerate(ts):           
    dt = datetime.datetime.fromtimestamp(t)	
    plankton = plankton_source[i][0][::-1,::]        
    plankton = np.ma.masked_where(isnan(plankton), plankton)
    plankton = plankton[np.min(latindexes):np.max(latindexes), np.min(lonindexes):np.max(lonindexes)]
    means.append(np.mean(np.ma.fix_invalid(plankton)))
    times.append(dt)
ax.plot(times, means)
plt.show()
```

Results are quite surprising: despite of expected presence of April's peak we should also see an extensive bloom at the summer time. Obviously, it is a good starting point to re-validate these calculations and maybe to compare with other datasets and time ranges.
